## Jolpica API Query

```python
import requests

url = 'https://api.jolpi.ca/ergast/f1/2024/driverstandings/'
response = requests.get(url)
```

```python
print('HTTP Status Code:', response.status_code)
print('Content Length:', len(response.content))
drivers = response.json()["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
print('Driver - Nationality - Points')
for d in drivers[:3]:
    print(d["Driver"]["familyName"], '-', d["Driver"]["nationality"], '-', d["points"])
```
